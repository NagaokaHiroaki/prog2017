```python
for x in range(1, 21):
    if x % 3 ==0:
        if x & 5 ==0:
            print('Fuzz Buzz')
        else:
            print('Fuzz')
    elif x % 5 ==0:
        print('Buzz')
    else:
        print(x)
```
